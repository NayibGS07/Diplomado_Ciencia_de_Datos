```python
# Función para dar formato al Título
def format_title(title, subtitle=None, subtitle_font_size=14):
        title = f'<b>{title}</b>'
        if not subtitle:
            return title
        subtitle = f'<span style="font-size: {subtitle_font_size}px;">{subtitle}</span>'
        return f'{title}<br>{subtitle}'
```

```python
fig = px.bar(df_exl, x = 'experience_level', y = 'Count', text = 'Count')
fig.update_layout(
    yaxis_title = 'Conteo',
    xaxis_title = 'Nivel de Experiencia',
    title_text = format_title('Grafica de Barras', 'Clase Diplomado'),
    template =  'plotly_white'
)
fig
```
