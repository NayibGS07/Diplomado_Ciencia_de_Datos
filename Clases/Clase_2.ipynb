```python
import pandas as pd
```

### Preparar Datos

```python
df = pd.read_csv('/content/salaries.csv')
df.head()
```

```python
df.shape
```

```python
# Ordenar DataFrame
df.sort_values('work_year')
```

```python
df.sort_values('experience_level')
```

```python
df.info()
```

```python
# Cambiar tipo de columna
from pandas.api.types import CategoricalDtype

cat_type = CategoricalDtype(categories = ['EN', 'MI', 'SE', 'EX'], ordered = True)
df['experience_level'] = df.experience_level.astype(cat_type)
df['work_year'] = df.work_year.astype(str)
df.head()
```

```python
df.info()
```

```python
df = df.sort_values(['work_year', 'experience_level', 'salary_in_usd'])
df.head()
```

```python
# Contar valores
df.value_counts('employment_type')
```

```python
# Porcentaje
df.value_counts('experience_level', normalize = True)
```

```python
# Diferentes Divisas
df.salary_currency.unique() 
```

```python
df['same_country'] = df.apply(lambda x: x.employee_residence == x.company_location, axis = 1)
df
```

```python
df.value_counts('same_country')
```

```python
fig = px.bar(df_1, x = 'experience_level', y = 'Count', text = 'Count')
fig.update_layout(
    xaxis_title = 'Nivel de Experiencia', # Título Eje X
    yaxis_title = 'Conteo', # Titulo Eje Y
    title_text = 'Grafica de Barras', # Título
    template = 'plotly_white', # Plantilla
    ) 
fig
```

```python
# Función para dar formato al Título
def format_title(title, subtitle=None, subtitle_font_size=14):
        title = f'<b>{title}</b>'
        if not subtitle:
            return title
        subtitle = f'<span style="font-size: {subtitle_font_size}px;">{subtitle}</span>'
        return f'{title}<br>{subtitle}'
```

```python
fig = px.bar(df_1, x = 'experience_level', y = 'Count', text = 'Count')
fig.update_layout(
    xaxis_title = 'Nivel de Experiencia', # Título Eje X
    yaxis_title = 'Conteo', # Titulo Eje Y
    title_text = format_title('Grafica de Barras', 'Utilizando Plotly'), # Título y Subtitulo
    template = 'plotly_white', # Plantilla
    ) 
fig
```

```python
df_1 = pd.DataFrame(df.groupby(['work_year']).experience_level.value_counts()).rename(columns = {'experience_level':'Count'}).reset_index()
df_1
```

```python
fig = px.bar(df_1, x = 'work_year', y = 'Count', text = 'Count', color = 'experience_level')
fig.update_layout(
    xaxis_title = 'Nivel de Experiencia', # Título Eje X
    yaxis_title = 'Conteo', # Titulo Eje Y
    title_text = format_title('Empleos por Año', 'Nivel de Experiencia'),
    template = 'plotly_white', # Plantilla
    ) 
fig
```

```python
fig = px.bar(df_1, x = 'work_year', y = 'Count', text = 'Count', color = 'experience_level', barmode = 'group')
fig.update_layout(
    xaxis_title = 'Nivel de Experiencia',
    yaxis_title = 'Conteo',
    title_text = format_title('Empleos por Año', 'Nivel de Experiencia'),
    template = 'plotly_white',
    ) 
fig
```

#### Gráfica de Puntos y Linea

https://plotly.com/python/line-and-scatter/

```python
df_1 = df.groupby(['work_year', 'company_size', 'experience_level'], observed = True).salary_in_usd.mean().reset_index()
df_1
```

```python
fig = px.scatter(df_1, x = 'work_year', y = 'salary_in_usd', color = 'experience_level', symbol = 'company_size')
fig.update_layout(
    xaxis_title = 'Año',
    yaxis_title = 'Salario Promedio',
    title_text = format_title('Salario Promedio por Año', 'Nivel de Experiencia'),
    template = 'plotly_white',
    ) 
fig
```

```python
df_1 = pd.DataFrame(df.groupby(['work_year']).remote_ratio.value_counts()).rename(columns = {'remote_ratio':'Count'}).reset_index()
df_1
```

```python
fig = px.line(df_1, x = 'work_year', y = 'Count', color = 'remote_ratio', markers = True)
fig.update_layout(
    xaxis_title = 'Año',
    yaxis_title = 'Puestos',
    title_text = format_title('Tipo de Trabajo por Año', 'Modalidad Remota'),
    template = 'plotly_white',
    ) 
fig
```

#### Histograma

https://plotly.com/python/histograms/

```python
fig = px.histogram(df, x = 'salary_in_usd')
fig.update_layout(
    xaxis_title = 'Salario',
    yaxis_title = '',
    title_text = format_title('Histograma', ''),
    template = 'plotly_white',
    ) 
fig
```

```python
fig = px.histogram(df, x = 'salary_in_usd', color = 'employment_type')
fig.update_layout(
    xaxis_title = 'Salario',
    yaxis_title = '',
    title_text = format_title('Histograma', ''),
    template = 'plotly_white',
    ) 
fig
```

```python
fig = px.histogram(df, x = 'salary_in_usd', color = 'employment_type', marginal = 'box')
fig.update_layout(
    xaxis_title = 'Salario',
    yaxis_title = '',
    title_text = format_title('Histograma', ''),
    template = 'plotly_white',
    ) 
fig
```

```python
fig = px.histogram(df, x = 'salary_in_usd', color = 'employment_type', marginal = 'box', facet_col = 'experience_level')
fig.update_layout(
    xaxis_title = 'Salario',
    yaxis_title = '',
    title_text = format_title('Histograma', ''),
    template = 'plotly_white',
    )
fig.for_each_annotation(lambda a: a.update(text=a.text.split('=')[-1])) 
fig
```

```python
fig = px.histogram(df, x = 'salary_in_usd', color = 'employment_type', marginal = 'box', facet_col = 'experience_level', facet_row = 'company_size',height = 900, width = 1300)
fig.update_layout(
    title_text = format_title('Histograma', ''),
    template = 'plotly_white',
    xaxis_title = '',
    xaxis2_title = '',
    xaxis3_title = '',
    xaxis4_title = 'Salario',
    )
fig.for_each_annotation(lambda a: a.update(text=a.text.split('=')[-1])) 
fig
```
